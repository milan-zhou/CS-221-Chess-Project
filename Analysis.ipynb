```python
import re
from matplotlib import pyplot
import numpy as np
pyplot.style.use('seaborn-deep')
files = [
    "/Users/milanzhou/Files/Stanford/2017-2018/Spring/CS 221/Project/logs_test/MCTS_1528445475.txt",
    "/Users/milanzhou/Files/Stanford/2017-2018/Spring/CS 221/Project/logs_test/minimax_1528169536.txt",
    "/Users/milanzhou/Files/Stanford/2017-2018/Spring/CS 221/Project/logs_test/alphabeta_1528169532.txt",
    "/Users/milanzhou/Files/Stanford/2017-2018/Spring/CS 221/Project/logs_test/mtd_1528190799.txt",
    "/Users/milanzhou/Files/Stanford/2017-2018/Spring/CS 221/Project/logs_value/stockfish_1527061042.txt"
]
scores = []
totals = []
nodes = []
for file in files:
    with open(file) as f:
        s = []
        t = []
        n = []
        for line in f.readlines():
            if re.search(">>.*: \((.*)\)", line):
                m = re.search(">>.*: \((.*)\)", line)
            
                score, tm, gm, avgNodes = m.groups(1)[0].split(", ")
                t.append(int(score))
                continue
            if re.search("STS/STS(\d+)", line):
                continue
            line = re.sub(" \{.*\},", "", line)
            num, nc, move, score = line.split(", ")
            s.append(int(score))
            n.append(int(nc))
        totals.append(t)
        nodes.append(n)
        scores.append(s)

for node in nodes:
    print (np.sum(np.array(node)) / 1400)
pyplot.figure(figsize=(10,8))
ax = pyplot.subplot(111)

a = -.3
bins = np.linspace(1, 14, 14)

for r in totals:
    print(r)
    ax.bar(bins + a,r,width=0.15,align='center')
    a += .15
l = ['MCTS', 'Minimax','Alpha-Beta','MTD(f)', 'Stockfish']
pyplot.xticks(bins)
pyplot.xlabel("STS Test #")
pyplot.ylabel("Score (out of 1000)")
pyplot.yticks(np.linspace(0,1000,11))
pyplot.title("Total Score on STS")
pyplot.savefig("Stockfish_Totals.png", dpi=300)
pyplot.show()

pyplot.figure(figsize=(10,8))
bins = np.linspace(0,11,12)
pyplot.hist(scores, bins,label=l, align='left')
pyplot.legend(loc='upper left')
pyplot.xlabel("Score")
pyplot.ylabel("Count")
pyplot.xticks(np.linspace(0,10,11))
pyplot.yticks(np.linspace(0,1500,16))
pyplot.title("STS Score Distribution")
font = {'family' : 'Arial',
        'size'   : 18}

pyplot.rc('font', **font)
pyplot.savefig("Stockfish_Scores.png", dpi=300)
pyplot.show()

pyplot.figure(figsize=(10,8))

bins = np.linspace(500, 1600, 110)

pyplot.hist(nodes, bins, label=l)
pyplot.legend(loc='upper right')
pyplot.xlabel("Number of Nodes")
pyplot.ylabel("Count")
pyplot.title("Number of Nodes Traversed")
font = {'family' : 'Arial',
        'size'   : 18}

pyplot.rc('font', **font)
pyplot.savefig("Stockfish_Nodes.png", dpi=300)
pyplot.show()
```
